Null Hypothesis: There is no relationship between both groups

Alternate Hypothesis: There is a relationship between both groups

# Question
Is the increase in food consumption from 2004 to 2014 statistically significant with increase in food production?

Is it the same for the top producing and consuming countries?


```python
# this function computes the correlation and the p-values for both groups
def calculate(group1, group2):
    corr, _ = pearsonr(group1, group2)
    print('Pearsons correlation: %.3f' % corr)
    print('p-value: %.3f' % _)
    if _ < 0.05:
        print('Reject null hypothesis >> Relationship')
    else:
        print('Fail to reject null hypothesis >> No Relationship')
```

```python
fprod = FoodProd.groupby('Year')['Value'].sum().tolist()

fconsmp = FoodSup.groupby('Year')['Value'].sum().tolist()
```

```python
# relationship between food production and food consumption in Africa
calculate(fprod, fconsmp)
```

```python
sns.scatterplot(fprod, fconsmp)
plt.xlabel('Yearly Production')
plt.ylabel('Yearly Consumption')
plt.title('Scatterplot Showing the Relationship Between Food Production and Consumption in Africa');
```
